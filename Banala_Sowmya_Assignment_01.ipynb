# Question 1

(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here
import random

# Generate 12 random numbers between 1 and 100
random_numbers = [random.randint(1, 100) for _ in range(12)]

# Sort the list in ascending order
ascending_sorted = sorted(random_numbers)

# Sort the list in descending order
descending_sorted = sorted(random_numbers, reverse=True)

# Calculate the average
average = sum(random_numbers) / len(random_numbers)

# Print the results
print("Random Numbers:", random_numbers)
print("Ascending Sorted:", ascending_sorted)
print("Descending Sorted:", descending_sorted)
print("Average:", average)
```

# Question 2
(25 points).

Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```python
# Your code here

import string

# Function to split a sentence into words while preserving punctuation
def split_sentence(sentence):
    words = []
    current_word = ''

    for char in sentence:
        if char.isalnum() or char in string.punctuation:
            current_word += char
        else:
            if current_word:
                words.append(current_word)
            current_word = ''

    if current_word:
        words.append(current_word)

    return words

# Function to count vowels in a sentence
def count_vowels(sentence):
    vowels = "aeiouAEIOU"
    vowel_count = {vowel: 0 for vowel in vowels}

    for char in sentence:
        if char in vowels:
            vowel_count[char.lower()] += 1

    return vowel_count

# Ask the user to enter two sentences
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# Make each sentence into a list of words
words1 = split_sentence(sentence1)
words2 = split_sentence(sentence2)

# Compare the two sentences and find common words
common_words = list(set(words1) & set(words2))

# Remove punctuation from the sentences
sentence1_no_punctuation = ''.join([char for char in sentence1 if char not in string.punctuation])
sentence2_no_punctuation = ''.join([char for char in sentence2 if char not in string.punctuation])

# Count the number of vowels in the sentences
vowel_count1 = count_vowels(sentence1)
vowel_count2 = count_vowels(sentence2)

# Print the results
print("\nWords in the first sentence:", words1)
print("Words in the second sentence:", words2)
print("Common words in both sentences:", common_words)
print("\nFirst sentence without punctuation:", sentence1_no_punctuation)
print("Second sentence without punctuation:", sentence2_no_punctuation)
print("\nVowel counts in the first sentence:", vowel_count1)
print("Vowel counts in the second sentence:", vowel_count2)
```

# Question 3

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```python
# Your code here
import math

# Function to calculate the volume of a sphere for a given radius
def calculate_volume(radius):
    return (4/3) * math.pi * radius**3

# Print the table header
print("r    V")

# Calculate and print the volume for each radius in the range 1-10
for radius in range(1, 11):
    volume = calculate_volume(radius)
    print(f"{radius}    {volume:.2f}")
```

```python
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('Assignment1_denton_housing.csv')

# Filter rows with the 'title_field' column equal to 'Occupied Housing Units'
occupied_units_data = data[data['title_field'] == 'Occupied Housing Units']

# Calculate the difference in Occupied Housing Units from year to year
occupied_units_data['difference'] = occupied_units_data['value'].diff()

# Create a new DataFrame to store the year ranges and differences
result_df = pd.DataFrame({
    'Year Range': occupied_units_data['year'].astype(str) + '-' + occupied_units_data['year'].shift(-1).astype(str),
    'Difference in Occupied Housing Units': occupied_units_data['difference']
})

# Sort the DataFrame by differences in ascending order
result_df = result_df.sort_values(by='Difference in Occupied Housing Units')

# Print the sorted results
print(result_df)
```

```python
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('Assignment1_denton_housing.csv')

# Filter rows for 'Vacant Housing Units' and 'Occupied Housing Units'
vacant_data = data[data['title_field'] == 'Vacant Housing Units']
occupied_data = data[data['title_field'] == 'Occupied Housing Units']

# Calculate the total housing units for each year
total_housing_units = vacant_data['value'] + occupied_data['value']

# Calculate the percentage of vacant and occupied housing units
percentage_vacant = (vacant_data['value'] / total_housing_units) * 100
percentage_occupied = (occupied_data['value'] / total_housing_units) * 100

# Create a DataFrame to store the results
result_df = pd.DataFrame({
    'Year': vacant_data['year'],
    'Percentage Vacant': percentage_vacant,
    'Percentage Occupied': percentage_occupied
})

# Print the results in the specified format
print("Year | Vacant Housing Units | Occupied Housing Units")
print("-" * 50)

# Iterate through the years and print the data
for year in range(2008, 2015):
    if year in result_df['Year'].tolist():
        row = result_df[result_df['Year'] == year].iloc[0]
        print(f"{year} | {row['Percentage Vacant']:.2f}% | {row['Percentage Occupied']:.2f}%")
    else:
        print(f"{year} | ----- | -----")
```
